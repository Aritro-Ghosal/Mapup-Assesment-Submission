### Question 1: Distance Matrix Calculation
Create a function named <b>calculate_distance_matrix</b> that takes the <b>dataset-3.csv</b> as input and generates a DataFrame representing distances between IDs.

The resulting DataFrame should have cumulative distances along known routes, with diagonal values set to 0. If distances between toll locations A to B and B to C are known, then the distance from A to C should be the sum of these distances. Ensure the matrix is symmetric, accounting for bidirectional distances between toll locations (i.e. A to B is equal to B to A).

```python
def calculate_distance_matrix(x):
    # Read the CSV file into a DataFrame
    data = pd.read_csv(x)

    # Create a pivot table to represent distances between IDs
    distance_matrix = pd.pivot_table(data, values='distance', index='id_start', columns='id_end', fill_value=0)

    # Make the matrix symmetric by filling missing values with transposed values
    distance_matrix = distance_matrix.add(distance_matrix.T, fill_value=0)

    # Iterate through the columns and update diagonal values to 0
    for col in distance_matrix.columns:
        distance_matrix.at[col, col] = 0

    return distance_matrix


# Example usage
csv_file_path = "C:\BOX\MapUp-Data-Assessment-F\datasets\dataset-3.csv"
resulting_distance_matrix = calculate_distance_matrix(csv_file_path)
resulting_distance_matrix
```
